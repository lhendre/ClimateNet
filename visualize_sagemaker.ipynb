```python
import xarray as xr
import netCDF4
```

```python
predictions = xr.open_dataset("predictions.nc")
predictions
```

```python
import cartopy.crs as ccrs
p = predictions["predictions"].isel(time=7).plot(
    subplot_kws=dict(projection=ccrs.Orthographic(-80, 35), facecolor="gray"),
    transform=ccrs.PlateCarree(),
)
p.axes.coastlines()
p.axes.gridlines()
```
